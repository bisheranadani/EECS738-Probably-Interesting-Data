K means implementation.
Using http://benalexkeen.com/k-means-clustering-in-python/ as a guide for implementation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

dfIris = pd.read_csv('Iris.csv')
df = dfIris.drop(['Id','SepalWidthCm','PetalLengthCm','Species'], axis=1)
```

```python
k = 3
centroids = {}
for i in range(1,k+1):
    centroids[i] = [np.random.uniform(df.iloc[:,0].min(),df.iloc[:,0].max()), np.random.uniform(df.iloc[:,1].min(), df.iloc[:,1].max())]

print(centroids)
```

```python
fig = plt.figure(figsize=(5, 5))
plt.scatter(df.iloc[:,0], df.iloc[:,1], color='k')
colmap = {1: 'r', 2: 'g', 3: 'b', 4: 'y', 5: 'k', 6: 'c', 7: 'm'}
for i in centroids.keys():
    plt.scatter(*centroids[i], color=colmap[i])
plt.show()
```

```python
def assignment(df, centroids):
    for i in centroids.keys():
        # sqrt((x1 - x2)^2 - (y1 - y2)^2)
        df['distance_from_{}'.format(i)] = (
            np.sqrt(
                (df.iloc[:,0] - centroids[i][0]) ** 2
                + (df.iloc[:,1] - centroids[i][1]) ** 2
            )
        )
    centroid_distance_cols = ['distance_from_{}'.format(i) for i in centroids.keys()]
    df['closest'] = df.loc[:, centroid_distance_cols].idxmin(axis=1)
    df['closest'] = df['closest'].map(lambda x: int(x.lstrip('distance_from_')))
    df['color'] = df['closest'].map(lambda x: colmap[x])
    return df

df = assignment(df, centroids)
print(df.head())
```

```python
fig = plt.figure()
plt.scatter(df.iloc[:,0], df.iloc[:,1], color=df['color'], alpha=0.5, edgecolor='k')
for i in centroids.keys():
    plt.scatter(*centroids[i], color=colmap[i])
# plt.xlim(0, 80)
# plt.ylim(0, 80)
plt.show()
```

```python
import copy

old_centroids = copy.deepcopy(centroids)

def update(k):
    for i in centroids.keys():
        centroids[i][0] = np.mean(df[df['closest'] == i].iloc[:,0])
        centroids[i][1] = np.mean(df[df['closest'] == i].iloc[:,1])
    return k

centroids = update(centroids)
```

```python
fig = plt.figure()
ax = plt.axes()
plt.scatter(df.iloc[:,0], df.iloc[:,1], color=df['color'], alpha=0.5, edgecolor='k')
for i in centroids.keys():
    plt.scatter(*centroids[i], color=colmap[i])
# plt.xlim(0, 80)
# plt.ylim(0, 80)
for i in old_centroids.keys():
    old_x = old_centroids[i][0]
    old_y = old_centroids[i][1]
    dx = (centroids[i][0] - old_centroids[i][0]) * 0.75
    dy = (centroids[i][1] - old_centroids[i][1]) * 0.75
    ax.arrow(old_x, old_y, dx, dy, head_width=0.1, head_length=0.2, fc=colmap[i], ec=colmap[i])
plt.show()
```

```python
def plot_scatter(df, centroids):
    fig = plt.figure()
    plt.scatter(df.iloc[:,0], df.iloc[:,1], color=df['color'], alpha=0.5, edgecolor='k')
    for i in centroids.keys():
        plt.scatter(*centroids[i], color=colmap[i], marker='*', s=150)    
    plt.show()
```

```python
import copy
def update(k):
    for i in centroids.keys():
        centroids[i][0] = np.mean(df[df['closest'] == i].iloc[:,0])
        centroids[i][1] = np.mean(df[df['closest'] == i].iloc[:,1])
    return k
```

```python
def assignment(df, centroids):
    for i in centroids.keys():
        # sqrt((x1 - x2)^2 - (y1 - y2)^2)
        df['distance_from_{}'.format(i)] = (
            np.sqrt(
                (df.iloc[:,0] - centroids[i][0]) ** 2
                + (df.iloc[:,1] - centroids[i][1]) ** 2
            )
        )
    centroid_distance_cols = ['distance_from_{}'.format(i) for i in centroids.keys()]
    df['closest'] = df.loc[:, centroid_distance_cols].idxmin(axis=1)
    df['closest'] = df['closest'].map(lambda x: int(x.lstrip('distance_from_')))
    df['color'] = df['closest'].map(lambda x: colmap[x])
    return df
```

```python
def kmeans(k, df):
    centroids = {}
    for i in range(1,k+1):
        centroids[i] = [np.random.uniform(df.iloc[:,0].min(),df.iloc[:,0].max()), np.random.uniform(df.iloc[:,1].min(), df.iloc[:,1].max())]
    
    df = assignment(df, centroids)
    centroids = update(centroids)
    
    while True:
        closest_centroids = df['closest'].copy(deep=True)
        centroids = update(centroids)
        df = assignment(df, centroids)
        if closest_centroids.equals(df['closest']):
            break
            
    return centroids
```
